Load the app reviews demo and explore with pandas.

```python
import pandas as pd

pd.set_option("display.max_columns", None)
print("pandas version:", pd.__version__)
```

```python
df = pd.read_csv("app_reviews_demo.csv")
df.head()
```

```python
#Prints a quick summary of the DataFrame
df.info()
```

```python
#Filters the DataFrame to only include rows where the rating is less than 3
low_ratings = df[df["rating"] < 3]
low_ratings
```

```python
#Mean rating of each app with dictionary of rating and average rating
df.groupby("app", as_index=False)["rating"].mean().rename(columns={"rating": "avg_rating"})
```

```python
#Meand of apps with rating < 3
df.loc[df["rating"] < 3].groupby("app")["rating"].agg(["mean", "count"])
```
