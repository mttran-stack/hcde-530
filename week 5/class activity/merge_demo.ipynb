# Week 5 — Merge demo (rebuilt)

Small example of **`pd.merge`**: joining two tables on a shared key (`app`). Use this if the original course merge file is missing.

- **`merge_demo_apps.csv`** — one row per app (includes **Notion**, not in the reviews sample).
- **`merge_demo_stats.csv`** — metrics per app (includes **Slack**, not in apps list) so you can see **inner / left / right / outer** behavior.

```python
import pandas as pd
from IPython.display import display

apps = pd.read_csv("merge_demo_apps.csv")
stats = pd.read_csv("merge_demo_stats.csv")
display(apps)
display(stats)
```

## Inner join — only keys in **both** tables

```python
inner = pd.merge(apps, stats, on="app", how="inner")
inner
```

## Left join — keep every row from **left** (`apps`)

```python
left = pd.merge(apps, stats, on="app", how="left")
left
```

## Right join — keep every row from **right** (`stats`)

```python
right = pd.merge(apps, stats, on="app", how="right")
right
```

## Outer join — all keys from either side

```python
outer = pd.merge(apps, stats, on="app", how="outer")
outer
```
